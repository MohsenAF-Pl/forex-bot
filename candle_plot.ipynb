```python
import pandas as pd
import utils
import instrument
```

```python
pair = "USD_JPY"
granularity = "H1"
ma_list = [20, 50]
i_pair = instrument.Instrument.get_instrument_by_name(pair)
```

```python
df = pd.read_pickle(utils.get_hist_data_filename(pair, granularity))
```

```python
print(df.head())
print("+++++++++++++++++++++++++++++++++++++++++")
print(df.shape)
```

```python
df.info()
```

```python
# we specify the columns that we don't want them to convert
non_cols = ['time', 'volume']
```

```python
# and here we specify the columns that we want to convert them numerical type
mod_cols = [x for x in df.columns if x not in non_cols]
mod_cols
```

```python
df[mod_cols] = df[mod_cols].apply(pd.to_numeric)
df.info()
```

```python
df.describe()
```

```python
import plotly.graph_objects as go
```

```python
# we make a data frame that contains 100 candles so we can plot them for practice

df_plot = df.iloc[-500:].copy()
```

```python
df_plot.shape
```

```python
fig = go.Figure()
fig.add_trace(go.Candlestick(
    x=df_plot.time, open=df_plot.mid_o, high=df_plot.mid_h, low=df_plot.mid_l, close=df_plot.mid_c,
    line=dict(width=1), opacity=1,
    increasing_fillcolor="#24A068",
    decreasing_fillcolor="#CC2E3C",
    increasing_line_color="#2EC886",
    decreasing_line_color="#FF3A4C"
))
fig.update_layout(width=1000, height=300,
    margin=dict(l=10, r=10, b=10, t=10),
    font=dict(size=10, color="#e1e1e1"),
    paper_bgcolor="#1e1e1e", plot_bgcolor="#1e1e1e"
)
fig.update_xaxes(
    gridcolor="#1f292f",
    showgrid=True, fixedrange=True, rangeslider=dict(visible=False)
)
fig.update_yaxes(
    gridcolor="#1f292f",
    showgrid=True
)
fig.show()
```

```python
df_ma = df[['time', 'mid_o', 'mid_h', 'mid_l', 'mid_c']].copy()
```

```python
df_ma.head()
```

```python
for ma in ma_list:
    df_ma[f"MA_{ma}"] = df_ma.mid_c.rolling(window=ma).mean()
```

```python
df_ma.dropna(inplace=True)
```

```python
df_ma.reset_index(drop=True, inplace=True)
```

```python
df_ma['DIFF'] = df_ma.MA_20 - df_ma.MA_50
df_ma.head()
```

```python
df_plot = df_ma.iloc[-500:].copy()
```

```python
df_ma['DIFF_PREV'] = df_ma.DIFF.shift(1)
```

```python
df_ma.head()
```

```python
def is_trade(row):
    if row.DIFF >= 0 and row.DIFF_PREV < 0:
        return 1
    if row.DIFF <= 0 and row.DIFF_PREV > 0:
        return -1
    return 0
```

```python
df_ma['IS_TRADE'] = df_ma.apply(is_trade, axis=1)
```

```python
df_ma.head()
```

```python
# we're extracting the rows that is_trade is true so the data we need come to the surface

df_trades = df_ma[df_ma.IS_TRADE!=0].copy()
df_ma.head()
```

```python
df_trades
```

```python
df_trades.shape
```

```python
fig = go.Figure()
fig.add_trace(go.Candlestick(
    x=df_plot.time, open=df_plot.mid_o, high=df_plot.mid_h, low=df_plot.mid_l, close=df_plot.mid_c,
    line=dict(width=1), opacity=1,
    increasing_fillcolor="#24A068",
    decreasing_fillcolor="#CC2E3C",
    increasing_line_color="#2EC886",
    decreasing_line_color="#FF3A4C"
))

for ma in ma_list:
    col = f"MA_{ma}"
    fig.add_trace(go.Scatter(x=df_plot.time, y=df_plot[col], line=dict(width=2),
        line_shape='spline', name=col
    ))

fig.update_layout(width=1000, height=300,
    margin=dict(l=10, r=10, b=10, t=10),
    font=dict(size=10, color="#e1e1e1"),
    paper_bgcolor="#1e1e1e", plot_bgcolor="#1e1e1e"
)
fig.update_xaxes(
    gridcolor="#1f292f",
    showgrid=True, fixedrange=True, rangeslider=dict(visible=False)
)
fig.update_yaxes(
    gridcolor="#1f292f",
    showgrid=True
)
fig.show()
```
